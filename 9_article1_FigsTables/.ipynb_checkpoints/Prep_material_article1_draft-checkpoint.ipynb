```python
# Converting the values in log scale - Need to convert the type to float
sen_cases_log = np.log(sen_cases.astype(float))
# If there are any infinity values, they need to be replaced by null values!
sen_cases_log.replace([np.inf, -np.inf], np.nan, inplace=True)
sen_cases_log.head()
```

```python
# Example
# https://www.tutorialspoint.com/matplotlib/matplotlib_bar_plot.htm
fig = plt.figure()
fig.set_size_inches(20,5)
ax = fig.add_axes([0,0,1,1])
#for
X = np.arange(len(sen_cases.columns)) # Number of parts to consider
x = len(sen_cases_log.index)

ax.bar(X + 0.0, sen_cases_log.iloc[0,:], color='tab:green', label='Case 1', width=0.25)
ax.bar(X + 0.33, sen_cases_log.iloc[1,:], color='tab:red', label='Case 2', width=0.25)
ax.bar(X + 0.66, sen_cases_log.iloc[2,:], color='tab:blue', label='Case 3', width=0.25)
major_ticks = np.arange(0, len(sen_cases_log.columns.values), 1)
ax.set_xticks(major_ticks)
labels = [item.get_text() for item in ax.get_xticklabels()]
labels = sen_cases_log.columns.values
ax.set_xticklabels(labels)
ax.legend(fontsize=14)
ax.set_ylabel('Log sensitivity values', fontsize=25)
ax.tick_params(axis='both', which='major', labelsize=15.5)
ax.tick_params(axis='both', which='minor', labelsize=15)
ax.set_xlabel('Parameters', fontsize=25)
#ax.axhline(y=-11.5,color='k',linestyle='-.')
ax.grid(b='True',which='both')

plt.savefig('log_sen_cases.jpg',dpi=300, bbox_inches="tight")
```
